```python
iris = datasets.load_iris()
df = pd.DataFrame(data = iris.data, columns = iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
data = df.values
nrow,ncol = data.shape
y = data[:,-1]
X = data[:,0:ncol-1]
```

```python
def likelihood(y, Z):
  mu = np.mean(Z, axis=0)
  cov = np.cov(Z, rowvar=False)
  prob = np.exp(-0.5*(y - mu).T@np.linalg.inv(cov)@(y - mu))/np.sqrt(np.linalg.det(2*np.pi*cov))
  return prob
```

```python
def Naive(X_train, y_train, X_test):
  classes = np.unique(y_train)
  valores = pd.DataFrame(data=np.zeros((X_test.shape[0],len(classes))),columns=classes)
  for i in range(len(classes)):
    elements = np.where(y_train == classes[i])[0]
    Z = X_train[elements]
    for j in range(X_test.shape[0]):
      x = X_test[j]
      pj = likelihood(x, Z)
      valores[classes[i]][j] = pj*len(elements)/X_train.shape[0]
  return valores
```
